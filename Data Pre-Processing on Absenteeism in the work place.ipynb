# .get_dummies( )
### Turn categorical data into dummy variables

In this case we want to make our 'reasons for absence" into 1 or 0. We can be certain that an individual has been absent from work becasue of one, and only on particular reason.


```python
reason_column_dummies = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_column_dummies
```

```python
reason_column_dummies['check'] = reason_column_dummies.sum(axis=1)
reason_column_dummies
```

## Check to see how many unique values are in the CHECK column

```python
reason_column_dummies['check'].unique()
```

## Drop the Check column now that we have confirmed the values in the rows are correct

```python
reason_column_dummies = reason_column_dummies.drop(['check'], axis=1)
```
